```python
# Step 3: Load Titanic CSV file

import pandas as pd

df = pd.read_csv('/content/train.csv')

# Display first 5 rows
print(df.head())
```

```python
# Step 4: Filter adults with fare > 30

filtered_df = df[(df['Age'] >= 18) & (df['Fare'] > 30)]

print(filtered_df.head())
```

```python
#step 5: Filter adults with fare > 30
filtered_df = df[(df['Age'] >= 18) & (df['Fare'] > 30)].copy()

# Family Size
filtered_df['family_size'] = (
    filtered_df['SibSp'] + filtered_df['Parch'] + 1
)

# Fare Per Person
filtered_df['fare_per_person'] = (
    filtered_df['Fare'] / filtered_df['family_size']
)

print(filtered_df[['Fare', 'family_size', 'fare_per_person']].head())
```

```python
# Step 6: Group By Analysis

group_analysis = filtered_df.groupby(['Pclass', 'Sex']).agg(
    average_fare=('Fare', 'mean'),
    survival_rate=('Survived', 'mean')
)

# Convert survival rate into percentage
group_analysis['survival_rate'] *= 100

print(group_analysis)
```
